```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
template = cv.imread('./template/pattern.png')
template_gray = cv.cvtColor(template, cv.COLOR_BGR2GRAY)
template_rgb = cv.cvtColor(template, cv.COLOR_BGR2RGB)

plt.figure(figsize=(4,4))
plt.imshow(template_rgb)
```

```python
image_paths = sorted([f for f in os.listdir('./images') if f.endswith('.png') or f.endswith('.jpg')])
images = {name: cv.imread(f'./images/{name}') for name in image_paths}
images_rgb = {name: cv.cvtColor(image, cv.COLOR_BGR2RGB) for name, image in images.items()}
images_gray = {name: cv.cvtColor(image, cv.COLOR_BGR2GRAY) for name, image in images.items()}

# Grilla con imagenes
plt.figure(figsize=(8, 8))
for i, name in enumerate(image_paths):
    plt.subplot(3, 3, i + 1)
    plt.imshow(images_rgb[name])
    plt.title(name)
plt.show()
```

```python
print(f'Template: {template_gray.shape}')
for name, img in images_gray.items():
    print(f'{name}: {img.shape}')
```
